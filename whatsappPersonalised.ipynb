## WhatsAppWeb Personalised Message

This Script finds all the Contacts on the whatsapp broadcast group and then generates a contact list from them

```python
from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from time import sleep
```

```python
web = webdriver.Firefox()

web.get('http://web.whatsapp.com')
```

```python
# Find the SideBar
side_bar = web.find_element_by_class_name('chatlist-panel')

# Find the BroadCast List with the Contact List
side_bar_contacts = side_bar.find_elements_by_class_name('emojitext')
```

```python
contact_list = [x for x in side_bar_contacts if x.text == 'Full Contact']

# Click the Given Group
contact_list[0].click()
```

```python
# Find the Chat Pane header to click, which would reveal the contacts
chat_header = web.find_elements_by_class_name('pane-chat-header')

chat_header[0].click()
```

```python
names = web.find_element_by_class_name('infinite-list').find_elements_by_class_name('chat-title')
names_text = [name.text for name in names]
names_text
```

```python
wait = WebDriverWait(web, 10)
```

```python
search_bar = web.find_element_by_id('input-chatlist-search')
```

```python
# Print the Given Names
sleep(5)

for name in names_text:
    
    search_bar = web.find_element_by_id('input-chatlist-search')
    search_bar.click()
    search_bar.send_keys(name)
    
    sleep(2)
    personal_chat = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'matched-text')))

    personal_chat.click()

    compose_box = web.find_element_by_class_name('pluggable-input-body')

    compose_box.send_keys('Merry (close to the deadline) Christmas {} 🎄 🎅'.format(name.split()[0]))

    send_button = web.find_element_by_class_name('compose-btn-send')

    send_button.click()
```
